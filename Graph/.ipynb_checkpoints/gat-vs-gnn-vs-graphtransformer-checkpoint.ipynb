```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.datasets import QM9
from torch_geometric.loader import DataLoader
from torch_geometric.nn import NNConv, GATConv, global_add_pool
from torch.utils.data import random_split
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, r2_score

# 加载 QM9 数据集
dset = QM9('.')
train_set, valid_set, test_set = random_split(dset, [110000, 10831, 10000])
trainloader = DataLoader(train_set, batch_size=32, shuffle=True)
validloader = DataLoader(valid_set, batch_size=32, shuffle=True)
testloader = DataLoader(test_set, batch_size=32, shuffle=True)

# 使用 GPU 如果可用
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
target_idx = 1  # 目标是极化率
```

```python
'''
Best trial:
  Value: 0.07092127000324702
  Params: 
    conv1_out_channels: 64
    conv2_out_channels: 48
    fc_out_channels: 32
    learning_rate: 0.0017282095473439127
'''
class GNNNet(nn.Module):
    def __init__(self, num_node_features, num_edge_features):
        super().__init__()
        conv1_net = nn.Sequential(nn.Linear(num_edge_features, 32),
                                  nn.ReLU(),
                                  nn.Linear(32, num_node_features * 64))
        conv2_net = nn.Sequential(nn.Linear(num_edge_features, 32),
                                  nn.ReLU(),
                                  nn.Linear(32, 64 * 48))
        self.conv1 = NNConv(num_node_features, 64, conv1_net)
        self.conv2 = NNConv(64, 48, conv2_net)
        self.fc_1 = nn.Linear(48, 32)
        self.out = nn.Linear(32, 1)

    def forward(self, data):
        batch, x, edge_index, edge_attr = data.batch, data.x, data.edge_index, data.edge_attr
        x = F.relu(self.conv1(x, edge_index, edge_attr))
        x = F.relu(self.conv2(x, edge_index, edge_attr))
        x = global_add_pool(x, batch)
        x = F.relu(self.fc_1(x))
        return self.out(x)
```

```python
'''
Best trial:
  Value: 0.05335596916737423
  Params: 
    conv1_out_channels: 32
    conv2_out_channels: 64
    num_heads1: 6
    num_heads2: 4
    fc_out_channels: 64
    learning_rate: 0.0012290853037013434
'''
class GATNet(nn.Module):
    def __init__(self, num_node_features):
        super().__init__()
        self.conv1 = GATConv(num_node_features, 32, heads=6)  
        self.conv2 = GATConv(32 * 6, 64, heads=4)  
        self.fc_1 = nn.Linear(64 * 4, 64)
        self.out = nn.Linear(64, 1)
    def forward(self, data):
        batch, x, edge_index = data.batch, data.x, data.edge_index
        x = F.relu(self.conv1(x, edge_index))
        x = F.relu(self.conv2(x, edge_index))
        x = global_add_pool(x, batch)  # 全局池化
        x = F.relu(self.fc_1(x))
        return self.out(x)
```

```python
'''
Best trial:
  Value: 0.8465700921738493
  Params: 
    hidden_dim: 32
    num_heads: 8
    num_layers: 4
    fc_out_channels: 96
    dropout: 0.1
    learning_rate: 0.003769923120618825
'''
class GraphTransformer(nn.Module):
    def __init__(self, num_node_features, num_edge_features, num_heads=8, num_layers=4, hidden_dim=32, fc_out_channels=96, dropout=0.1):
        super(GraphTransformer, self).__init__()
        self.node_encoder = nn.Linear(num_node_features, hidden_dim)
        self.edge_encoder = nn.Linear(num_edge_features, hidden_dim)
        self.pos_encoder = nn.Linear(1, hidden_dim)  # 用于对节点进行位置编码
        self.dropout = nn.Dropout(dropout)
        # Transformer encoder layers
        encoder_layer = nn.TransformerEncoderLayer(d_model=hidden_dim, nhead=num_heads, dim_feedforward=hidden_dim*2, dropout=dropout)
        self.transformer_encoder = nn.TransformerEncoder(encoder_layer, num_layers)
        # Pooling and final output layers
        self.fc1 = nn.Linear(hidden_dim, fc_out_channels)
        self.fc2 = nn.Linear(fc_out_channels, 1)
    def forward(self, data):
        # 获取图数据的节点特征和边特征
        x, edge_index, edge_attr, batch = data.x, data.edge_index, data.edge_attr, data.batch
        
        # 进行节点特征编码
        x = self.node_encoder(x)
        # 获取节点数
        num_nodes = x.size(0)
        # 生成位置编码 (我们用节点索引进行简单位置编码)
        positions = torch.arange(0, num_nodes, dtype=torch.float32).unsqueeze(1).to(x.device)
        position_encoded = self.pos_encoder(positions)
        
        # 将位置编码和节点特征加在一起
        x = x + position_encoded
        x = self.dropout(x)
        # Transformer expects input shape (seq_len, batch, features), so we need to reshape
        x = x.unsqueeze(1)  # (num_nodes, 1, hidden_dim)
        # Transformer Encoder
        x = self.transformer_encoder(x)  # (num_nodes, 1, hidden_dim)
        x = x.squeeze(1)  # (num_nodes, hidden_dim)
        # 全局池化，将节点特征汇总为图的全局表示
        x = global_add_pool(x, batch)
        # MLP for prediction
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x
```

```python
def train_model(model, optimizer, trainloader, validloader, epochs=4):
    train_losses, val_losses = [], []
    for epoch in range(epochs):
        model.train()
        epoch_loss = 0
        total_graphs = 0
        for batch in trainloader:
            batch.to(device)
            optimizer.zero_grad()
            output = model(batch)
            loss = F.mse_loss(output, batch.y[:, target_idx].unsqueeze(1))
            loss.backward()
            optimizer.step()
            epoch_loss += loss.item()
            total_graphs += batch.num_graphs
        train_avg_loss = epoch_loss / total_graphs
        train_losses.append(train_avg_loss)
        # 验证
        model.eval()
        val_loss = 0
        total_graphs = 0
        with torch.no_grad():
            for batch in validloader:
                batch.to(device)
                output = model(batch)
                loss = F.mse_loss(output, batch.y[:, target_idx].unsqueeze(1))
                val_loss += loss.item()
                total_graphs += batch.num_graphs
        val_avg_loss = val_loss / total_graphs
        val_losses.append(val_avg_loss)
        print(f"Epoch {epoch + 1}/{epochs} | Train Loss: {train_avg_loss:.4f} | Val Loss: {val_avg_loss:.4f}")
    
    return train_losses, val_losses
# 初始化三个模型及优化器
gnn_net = GNNNet(11, 4).to(device)
gat_net = GATNet(11).to(device)
graph_transformer_net = GraphTransformer(11, 4).to(device)
gnn_optimizer = torch.optim.Adam(gnn_net.parameters(), lr=0.0017282095473439127)
gat_optimizer = torch.optim.Adam(gat_net.parameters(), lr=0.0012290853037013434)
gt_optimizer = torch.optim.Adam(graph_transformer_net.parameters(), lr=0.003769923120618825)
# 训练每个模型
gnn_train_losses, gnn_val_losses = train_model(gnn_net, gnn_optimizer, trainloader, validloader, epochs=10)
gat_train_losses, gat_val_losses = train_model(gat_net, gat_optimizer, trainloader, validloader, epochs=10)
gt_train_losses, gt_val_losses = train_model(graph_transformer_net, gt_optimizer, trainloader, validloader, epochs=10)

# 绘制训练损失对比
plt.figure(figsize=(10, 5))

# GNN 训练损失
plt.plot(gnn_train_losses, label='GNN Train Loss', linestyle='--', marker='o')
# GAT 训练损失
plt.plot(gat_train_losses, label='GAT Train Loss', linestyle='--', marker='x')
# Graph Transformer 训练损失
plt.plot(gt_train_losses, label='Graph Transformer Train Loss', linestyle='--', marker='s')

plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training Loss Comparison: GNN, GAT, Graph Transformer')
plt.legend()
plt.show()

# 绘制验证损失对比
plt.figure(figsize=(10, 5))

# GNN 验证损失
plt.plot(gnn_val_losses, label='GNN Val Loss', linestyle='-', marker='o')
# GAT 验证损失
plt.plot(gat_val_losses, label='GAT Val Loss', linestyle='-', marker='x')
# Graph Transformer 验证损失
plt.plot(gt_val_losses, label='Graph Transformer Val Loss', linestyle='-', marker='s')

plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Validation Loss Comparison: GNN, GAT, Graph Transformer')
plt.legend()
plt.show()
```

```python
def test_model(model, testloader):
    model.eval()
    predictions = []
    real = []
    with torch.no_grad():
        for batch in testloader:
            output = model(batch.to(device))
            predictions.append(output.detach().cpu().numpy())
            real.append(batch.y[:, target_idx].detach().cpu().numpy())
    
    predictions = np.concatenate(predictions)
    real = np.concatenate(real)
    
    plt.scatter(real[:500], predictions[:500], alpha=0.7)
    plt.ylabel('Predicted isotropic polarizability')
    plt.xlabel('Real isotropic polarizability')
    plt.show()
# 测试并可视化每个模型的结果
print("GNN Test Results")
test_model(gnn_net, testloader)
print("GAT Test Results")
test_model(gat_net, testloader)
print("Graph Transformer Test Results")
test_model(graph_transformer_net, testloader)
```
